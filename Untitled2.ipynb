```python
import csv
import json
import numpy as np
import pandas as pd
import keras
import time
from keras import regularizers
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit
from matplotlib import pyplot
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import tensorflow as tf
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from keras.layers import Dropout
from sklearn.utils import shuffle

start = time.time()
weights=['bold']

dataframe = pd.read_csv("C:/Users/po19996/OneDrive - University of Bristol/Desktop/dataset01/lightpath.csv")
dataframe = shuffle(dataframe)
dataset = dataframe.values
X = dataset[:,1:9]
Y = dataset[:,9:12]
print("X.shape is ", X.shape)
print("Y.shape is ", Y.shape)

seed = 10
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X, Y, test_size=0.98, random_state = seed)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test,test_size = 0.5)
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)


scalerX = MinMaxScaler()
scalerX.fit(X_train)
scalered_X = scalerX.transform(X_train)
scalerY = MinMaxScaler()
scalerY.fit(Y_train)
scalered_Y = scalerY.transform(Y_train)
scalered_X_val = scalerX.transform(X_val)
scalered_Y_val = scalerY.transform(Y_val)
scalered_X_test = scalerX.transform(X_test)
scalered_Y_test = scalerY.transform(Y_test)


model = Sequential()
model.add(Dense(units= 24, activation='relu', input_dim = 8, kernel_regularizer=regularizers.l2(0.000001)))
model.add(Dense(units= 24, activation='relu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(Dense(3, activation='sigmoid',kernel_regularizer=regularizers.l2(0.00001)))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
print(model.summary())


hist = model.fit(scalered_X, scalered_Y,validation_data=(scalered_X_val, scalered_Y_val), epochs=150, batch_size=20,verbose=1)
prediction_Y = scalerY.inverse_transform(model.predict(scalered_X_test))




plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()


plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()


MSE = mean_squared_error(Y_test,prediction_Y)
print('Test MSE: %.4f MSE' % (MSE))
MAE = mean_absolute_error(Y_test,prediction_Y)
print('Test MAE: %.4f MAE' % (MAE))
R2 = r2_score(Y_test,prediction_Y)
print('Test R2: %.4f R2' % (R2))

# save model to file
#model.save('C:/Users/po19996/OneDrive - University of Bristol/Desktop/dataset01/model.h5')
```
